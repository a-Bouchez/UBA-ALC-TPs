## Definiciones de funciones para el TP

(les debería ser más fácil de corregir ordenado así)

```python
import numpy as np
import scipy.linalg

def construye_adyacencia(D,m):
    """
    Función que construye la matriz de adyacencia del grafo de museos
    D matriz de distancias, m cantidad de links por nodo
    Retorna la matriz de adyacencia como un numpy.
    """
    D = D.copy()
    l = [] # Lista para guardar las filas
    for fila in D: # recorriendo las filas, anexamos vectores lógicos
        l.append(fila<=fila[np.argsort(fila)[m]] ) # En realidad, elegimos todos los nodos que estén a una distancia menor o igual a la del m-esimo más cercano
    A = np.asarray(l).astype(int) # Convertimos a entero
    np.fill_diagonal(A,0) # Borramos diagonal para eliminar autolinks
    return(A)

def elim_gaussiana(A):
  """
  Realiza la eliminación gaussiana para triangularizar la matriz A.
  Devuelve la matriz L (triangular inferior) y U (triangular superior) resultantes.
  También devuelve la cantidad de operaciones realizadas.
  """
  cant_op = 0
  m=A.shape[0]
  n=A.shape[1]
  Ac = np.copy(A)

  if m!=n:
    print('Matriz no cuadrada')
    return
  
  for k in range(m-1): # para el k-ésimo paso
    for i in range(k+1, n): # para cada fila
      coeficiente = Ac[i,k] / Ac[k,k]
      cant_op += 1
      
      for j in range(m): # para cada columna

        if k < j:
          Ac[i,j] -= coeficiente * Ac[k,j] # triangularización
          cant_op += 2

        if k == j: Ac[i,j] = coeficiente # guardamos el coeficiente
      
    #print("La matriz A^(" + str(k + 1) + ") es:\n", Ac)
  
  #print("Matriz final con L en donde van los ceros:\n", A)
  L = np.tril(Ac,-1) + np.eye(A.shape[0])
  U = np.triu(Ac)

  return L, U, cant_op

def calculaLU(matriz):
    """
    Función para calcular la descomposición LU de una matriz cuadrada.
    matriz: Matriz cuadrada a descomponer
    Retorna: L, U.
    """
    L, U, _ = elim_gaussiana(matriz)
    return L, U

def calcular_numero_condición(A, p = None):
    """
    Función para calcular el número de condición de norma p de la matriz A.
    A: Matriz a calcular el número de condición.
    p: Número de norma a calcular.
    Retorna: Número de condición de norma p de la matriz A.
    """
    # Busco X / AX = I
    I = np.eye(A.shape[0]) # Matriz identidad
    L, U, = calculaLU(A) # Calculo la LU de a, luego tengo LUX = I
    Y = scipy.linalg.solve_triangular(L,I,lower=True) # Resuelvo el sistema LY = I con Y = UX
    X = scipy.linalg.solve_triangular(U,Y) # Resuelvo el sistema UX = Y
    return np.linalg.norm(A, p) * np.linalg.norm(X, p) # Devuelvo el número de condición de norma p de la matriz A
    

def calcula_matriz_C(A):
    """
    Función para calcular la matriz de trancisiones C.
    Usamos la ecuación C = A^T @ K^(-1).
    A: Matriz de adyacencia
    K: Matriz de grado (de salida), tiene en su diagonal la suma por filas de A
    Retorna la matriz C
    """
    Atrans = np.transpose(A) # Transponemos la matriz A
    
    # Calcula inversa de la matriz K, que tiene en su diagonal la suma por filas de A
    Kinv = np.zeros(A.shape) # Inicializamos Kinv con ceros
    for i in range(A.shape[0]): # Recorremos las filas de A
        suma = np.sum(A[i,:]) # Suma de la fila i-ésima de A
        if suma != 0: # Si la suma no es cero, asignamos el inverso a la diagonal de Kinv
            Kinv[i,i] = 1/suma
        else: # Si la suma es cero, asignamos cero a la diagonal de Kinv
            Kinv[i,i] = 0

    C = Atrans @ Kinv # Calcula C multiplicando Kinv y A
    return C

def calcula_pagerank(A,alfa):
    """
    Función para calcular PageRank usando LU.
    A: Matriz de adyacencia.
    alfa: Coeficiente de damping.
    Retorna: Un vector p con los coeficientes de PageRank de cada museo.
    """
    if alfa <= 0 or alfa >= 1: # Verificamos que alfa esté entre 0 y 1
        raise ValueError("Alfa debe estar entre 0 y 1 (exclusive).")
    
    C = calcula_matriz_C(A) # Calculamos la matriz de transiciones C
    N = A.shape[0] # Obtenemos el número de museos N a partir de la estructura de la matriz A
    
    M = N/alfa * (np.eye(N) - ((1-alfa) * C)) # Calculamos la matriz M con la ecuación del PDF
    b = np.ones(N) # Vector de 1s a igualar | en el comentario anterior sugería que multiplicáramos por alfa/N, pero no es necesario pues ya multiplicamos a M por N/alfa
    
    L, U = calculaLU(M) # Calculamos descomposición LU a partir de C y d
    Up = scipy.linalg.solve_triangular(L,b,lower=True) # Primera inversión usando L
    p = scipy.linalg.solve_triangular(U,Up) # Segunda inversión usando U
    return p
```

# Resolución del TP

Aquí empieza la aventura... ¡diviertanse y consulten lo que necesiten!

## Punto 1:

Empezamos con: $$p = (1 - \alpha)Cp + \frac{\alpha}{N}\cdot\textbf{1}$$
Queremos sacar todo lo que tenga $p$ del mismo lado. Restamos $(1 - \alpha)Cp$ a ambos lados: $$p - (1 - \alpha)Cp = \frac{\alpha}{N}\cdot\textbf{1}  \ \ \ \ \ (2)$$
Factorizamos $p$ en el lado izquierdo: $$(I - (1 - \alpha)C)p = \frac{\alpha}{N} \cdot \textbf{1}$$
Ahora multiplicamos ambos lados de la ecuación (2) por $\frac{N}{\alpha}$, que es una constante escalar: $$\frac{N}{\alpha}(I - (1 - \alpha)C)p = \frac{N}{\alpha} \frac{\alpha}{N}\cdot \textbf{1}$$
Obteniendo así: $$\underbrace{\frac{N}{\alpha}(I - (1 - \alpha)C)}_{M} p = \textbf{1} $$

Luego, $M p = \textbf{1}$.

## Punto 2:

Para que el sistema tenga solución única, la matriz $M$ debe ser inversible.

Veamos que se cumple que la matriz $M$ es inversible para $0 < \alpha < 1$ con una demostración.

Vamos a probar un lema primero.

**Lema I:** Sea $\tilde{C}$ una matriz cuadrada. Si $\|\tilde{C}\|_1 < 1$, entonces $(I - \tilde{C})$ es inversible.  
Vamos a hacer la demostración por construcción. Supongamos que $\|\tilde{C}\|_1 < 1$.

Sabemos que una matriz $A$ es inversible si y solo si $A x = \textbf{0}$ tiene como única solución $x = \textbf{0}$.  

Quiero ver que, $(I - \tilde{C}) x = \textbf{0}$ tiene como única solución $x = \textbf{0}$.

$$(I - \tilde{C}) x = \textbf{0}$$
$$I x - \tilde{C}x = \textbf{0}$$
$$x = \tilde{C}x$$

Luego, tomamos $\|\cdot\|_1$ a ambos lados, pues sabemos que $\|x\|_1 = 0 \iff x = \textbf{0}$.

$$\|x\|_1 = \|\tilde{C}x\|_1 \leq_{\text{propiedad submultiplicativa}} \|\tilde{C}\|_1 \|x\|_1$$
$$\|x\|_1 \leq \|\tilde{C}\|_1 \|x\|_1$$
$$\|x\|_1 - \|\tilde{C}\|_1 \|x\|_1 \leq 0$$
$$(1 - \|\tilde{C}\|_1) \|x\|_1 \leq 0$$

Por hipótesis, sabemos que $\|\tilde{C}\|_1 < 1$, y al ser norma vale que $\|\tilde{C}\|_1 \geq 0$. Luego, $0 < (1 - \|\tilde{C}\|_1) \leq 1$.  
A su vez, $\|x\|_1 \geq 0$.

Luego, para que valga la desigualdad que hicimos con $\iff$ arriba, es obligatorio que $\|x\|_1 = 0$.  
O, lo que es lo mismo, $x = \textbf{0}$.

Finalmente, concluimos que $(I - \tilde{C}) x = \textbf{0}$ tiene como única solución $x = \textbf{0}$, como se quería probar.  
Esto quiere decir que $(I - \tilde{C})$ es inversible.


Ahora, para usar el **Lema I**, argumentemos porqué vale que $\|(1 - \alpha)C\|_1 < 1$.  
Sabemos que $C$ es una matriz _estocástica_, es decir, **sus columnas suman $1$**; esto implica que $\|C\|_1 = 1$.  
Luego, vale que $\|(1 - \alpha)C\|_1 = |1 - \alpha| \cdot \|C\|_1 = |1 - \alpha|$.  
Como $0 < \alpha < 1$, el módulo no me afecta en nada, es decir, $|1 - \alpha| = 1 - \alpha$.  

Luego, $$0 < 1 - \alpha < 1$$

Finalmente, vale que $\|(1 - \alpha)C\|_1 < 1$. Por lo que la matriz $(I - (1 - \alpha)C)$ es inversible (por el **Lema I**).

Por último, sabemos que $\frac{N}{\alpha}(I - (1 - \alpha)C)$ también es inversible, esto se debe a que estamos multiplicando $\frac{N}{\alpha} \neq 0$,  
y multiplicar una matriz inversible por una constante $\neq 0$ no afecta su inversibilidad, pues no cambia el rango de la matriz.

## Punto 4:

Sea $v_k$ la cantidad de visitantes que tuvo cada museo en el k-ésimo paso (misma definición que el PDF del enunciado).  
Sabemos que $w = \sum_{k=0}^{r-1} v_k$ por definición.

Luego, $w = \sum_{k=0}^{r-1} v_k =_{\text{def. de } v_k \text{ del enunciado}} \sum_{k=0}^{r-1} C^k v_0 =_{\text{def. de } B} B v_0$

Notar que $v_0 = v$ por definición del enunciado, luego $w = B v$.  
Finalmente, asumiendo $B$ inversible _(un docente dijo en el labo que podemos asumir $B$ inversible)_.
$$v = B^{-1} w$$

## Punto 5:

```python
def calcula_matriz_C_continua(D):
    """
    Función para calcular la matriz de trancisiones C
    D: Matriz de distancias
    Retorna la matriz C en versión continua
    """
    D = D.copy()
    F = 1/D
    np.fill_diagonal(F,0)
    # Calcula inversa de la matriz K, que tiene en su diagonal la suma por filas de F
    Kinv = np.zeros(F.shape) # Inicializamos Kinv con ceros
    for i in range(F.shape[0]): # Recorremos las filas de F
        suma = np.sum(F[i,:]) # Suma de la fila i-ésima de F
        if suma != 0: # Si la suma no es cero, asignamos el inverso a la diagonal de Kinv
            Kinv[i,i] = 1/suma
        else: # Si la suma es cero, asignamos cero a la diagonal de Kinv
            Kinv[i,i] = 0
    C = Kinv@F # Calcula C multiplicando Kinv y F
    return C

def calcula_B(C,cantidad_de_visitas):
    """
    Recibe la matriz C de transiciones, y calcula la matriz B que representa la relación entre el total de visitas y el número inicial de visitantes suponiendo que cada visitante realizó cantidad_de_visitas pasos
    C: Matriz de transiciones
    cantidad_de_visitas: Cantidad de pasos en la red dado por los visitantes. Indicado como r en el enunciado
    Retorna una matriz B que vincula la cantidad de visitas w con la cantidad de primeras visitas v
    """
    B = np.eye(C.shape[0])
    C_potencia = np.eye(C.shape[0])  # Inicializamos C^k como la identidad
    for _ in range(cantidad_de_visitas - 1):
        # Sumamos las matrices de transición para cada cantidad de pasos
        C_potencia = C_potencia @ C  # Calculamos C^k
        B = B + C_potencia  # Sumamos C^k a B
    return B
```

```python
with open('visitas.txt', 'r') as file:
  w = np.array([int(line.strip()) for line in file.readlines()])

cantidad_de_visitas = len(w) # Cantidad de visitas
C = calcula_matriz_C_continua(D) # Calculamos la matriz de transiciones C en versión continua
B = calcula_B(C, cantidad_de_visitas) # Calculamos la matriz B a partir de C y la cantidad de visitas

L, U = calculaLU(B) # Calculamos descomposición LU de B para resolver el sistema Bv = w
Uv = scipy.linalg.solve_triangular(L,w,lower=True) # Primera inversión usando L
v = scipy.linalg.solve_triangular(U,Uv) # Segunda inversión usando U

print(v)
```

## Punto 6:

El enunciado está mal. $\tilde{w}$ y $\tilde{v}$ son el $w$ y $v$ originales, más un $\delta w$ y $\delta v$ _(ruido)_ respectivamente.  
Es decir, sean $v, w$ los valores reales para el total de visitas y el total de primeras visitas respectivamente, se tiene que:  
$$\tilde{w} = w + \delta w$$
$$\tilde{v} = v + \delta v$$

Queremos calcular el número de condición de $B$ y estimar la cota para el error _(relativo)_ de estimación de $v$, sabiendo que $\frac{\|w-\tilde{w}\|_1}{\|w\|_1} = 0.05$ pues el número total de visitantes tiene un error del $5$%.

Para ello, calculamos el número de condición $1$ de $B$ usando NumPy.

```python
Cond_B_1 = calcular_numero_condición(B, 1)
print("Número de condición (normal 1) de B:", Cond_B_1)

print("Número de condición (normal 1) de B multiplicado por el error relativo de w:", Cond_B_1 * 0.05)
```

Luego, $Cond_1(B) \approx 1795.53$.

Por lo tanto, la cota estimada para el error de estimación de $v$ es,
$$\frac{\|v-\tilde{v}\|_1}{\|v\|_1} \leq  Cond_1(B) \frac{\|w-\tilde{w}\|_1}{\|w\|_1}$$
$$\frac{\|v-\tilde{v}\|_1}{\|v\|_1} \leq  1795.53 \cdot 0.05$$
$$\frac{\|v-\tilde{v}\|_1}{\|v\|_1} \leq  89.78$$
